```python
from pandas import read_csv
transactions = read_csv("C:\\Users\\palej\\Desktop\\FIUBA\\Organizacion de datos\\TP\\transactions_train.csv")
```

```python
def tipo_transferencia(numero) -> str:
    if numero == 0:
        return "Legal"
    else:
        return "Fraude"
    
def unicidad_correo(string) -> str:
    lista = ["yahoo", "gmail", "hotmail", "outlook", "live"]
    for correo in lista:
        match string.lower().find(correo):
            case -1:
                continue
            case _:
                return f"{correo}.com"
    return string
```

```python
analisis = transactions[["target", "feat_card_6", "payer_email_domain"]]
```

```python
analisis = analisis.rename(columns={"target":"Transaccion", "feat_card_6":"Banco", "payer_email_domain":"Correo"})
```

```python
analisis = analisis.fillna("Otro")
```

```python
analisis.Transaccion = analisis.Transaccion.map(tipo_transferencia)
```

```python
analisis_banco = analisis.pivot_table(values="Correo", index="Banco", columns="Transaccion", aggfunc="count")
```

```python
analisis_banco = analisis_banco.fillna(0)
```

```python
analisis_banco["Promedio fraude en %"] = round(analisis_banco.Fraude * 100 / (analisis_banco.Fraude + analisis_banco.Legal), 2)
```

```python
analisis_banco
```

```python
analisis_correo = analisis.pivot_table(values="Banco", index="Correo", columns="Transaccion", aggfunc="count")
```

```python
analisis_correo = analisis_correo.fillna(0)
```

```python
analisis_correo = analisis_correo.reset_index()
```

```python
analisis_correo["Unicidad de correo"] = analisis_correo.Correo.map(unicidad_correo)
```

```python
analisis_correo = analisis_correo.groupby("Unicidad de correo").sum()
```

```python
analisis_correo.pop("Correo")
```

```python
analisis_correo["Promedio fraude en %"] = round(analisis_correo.Fraude * 100 / (analisis_correo.Fraude + analisis_correo.Legal), 2)
```
