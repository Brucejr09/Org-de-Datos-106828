```python
variables_categoricas = ["product_category", "feat_card_4", "feat_card_6", "payer_email_domain", "receiver_email_domain", "C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9"]

for variable in variables_categoricas:
    data_set[variable] = data_set[variable].fillna("NULO")
```

```python
def diferencia_listas(principal, diferencia):
    for elemento in diferencia:
        principal.remove(elemento)

    return principal

variables_numericas = diferencia_listas(data_set.nunique().keys().to_list(), variables_categoricas)

for variable in variables_numericas:
    data_set[variable] = data_set[variable].fillna(data_set[variable].mean())
```

```python
target = transactions.target
target
```

```python
from sklearn.model_selection import train_test_split

data_set_train, data_set_validation, target_train, target_validation = train_test_split(data_set, target, random_state=11)
```
