Manejo de NaN

```python
variables_categoricas = ["product_category", "feat_card_4", "feat_card_6", "payer_email_domain", "receiver_email_domain"]

for variable in variables_categoricas:
    data_set_train[variable] = data_set_train[variable].fillna("NULO")
    data_set_validation[variable] = data_set_validation[variable].fillna("NULO")
```

```python
def diferencia_listas(principal, diferencia):
    for elemento in diferencia:
        principal.remove(elemento)

    return principal

variables_numericas = diferencia_listas(data_set_train.nunique().keys().to_list(), variables_categoricas)

for variable in variables_numericas:
    data_set_train[variable] = data_set_train[variable].fillna(data_set_train[variable].mean())
    data_set_validation[variable] = data_set_validation[variable].fillna(data_set_validation[variable].mean())
```

Encoding de las variables categoricas

```python
cantidades_categoricas = {}

for variable in variables_categoricas:
    cantidades_categoricas[variable] = data_set_train[variable].nunique()

cantidades_categoricas
```

```python
import category_encoders as ce

for variable in variables_categoricas:
    target_encoder = ce.TargetEncoder(cols=variable)
    data_set_train[variable] = target_encoder.fit_transform(data_set_train[variable], data_set_train.feat_card_3)
    data_set_validation[variable] = target_encoder.transform(data_set_validation[variable], data_set_validation.feat_card_3)
```

Normalizacion

```python
def normalizar(numero) -> float:
    if numero != 0:
        return 1/numero
    return 0

for variable in data_set_train.columns.to_list():
    data_set_train[variable] = data_set_train[variable].map(normalizar)
    data_set_validation[variable] = data_set_validation[variable].map(normalizar)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
log_reg = LogisticRegression()
```

```python
parametros = [
    {"penalty": ["l2"]},
    {"C": [2.0]},
    {"fit_intercept": [True]},
    {"solver": ['saga']}
]
```

```python
hiper = GridSearchCV(log_reg, param_grid=parametros, n_jobs=5, scoring="roc_auc")
```

```python
hiper.fit(data_set_train, target_train)
```

```python
hiper.best_params_
```

```python
modelo = hiper.best_estimator_
```

```python
modelo.fit(data_set_train, target_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
```

```python
target_train_predict = modelo.predict(data_set_validation)
```

```python
target_validation = target_validation.astype(int)
```

```python
matriz_confusion = confusion_matrix(target_validation, target_train_predict)
matriz_confusion
```
