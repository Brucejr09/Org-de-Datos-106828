```python
from sklearn.model_selection import train_test_split

data_set_train_fraude, data_set_validation_fraude, target_train_fraude, target_validation_fraude = train_test_split(transactions[transactions.target == 1], target[target == 1], test_size=0.2, random_state=11)
```

```python
data_set_train = concat([data_set_train_legal, data_set_train_fraude])
data_set_validation = concat([data_set_validation_legal, data_set_validation_fraude])
target_train = concat([target_train_legal, target_train_fraude])
target_validation = concat([target_validation_legal, target_validation_fraude])
```

Aislar columnas innecesarias

```python
data_set_train = data_set_train.drop(["Unnamed: 0", "target", "Id"], axis=1)
data_set_validation= data_set_validation.drop(["Unnamed: 0", "target", "Id"], axis=1)
```

Manejo de NaN

```python
variables_categoricas = ["product_category", "feat_card_4", "feat_card_6", "payer_email_domain", "receiver_email_domain", "C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9"]

for variable in variables_categoricas:
    data_set_train[variable] = data_set_train[variable].fillna("NULO")
    data_set_validation[variable] = data_set_validation[variable].fillna("NULO")
```

```python
def diferencia_listas(principal, diferencia):
    for elemento in diferencia:
        principal.remove(elemento)

    return principal

variables_numericas = diferencia_listas(data_set_train.nunique().keys().to_list(), variables_categoricas)

for variable in variables_numericas:
    data_set_train[variable] = data_set_train[variable].fillna(data_set_train[variable].mean())
    data_set_validation[variable] = data_set_validation[variable].fillna(data_set_validation[variable].mean())
```
