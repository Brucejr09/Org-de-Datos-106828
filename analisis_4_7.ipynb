```python
import pandas as pd

identity = pd.read_csv("C:\\Users\\palej\\Desktop\\FIUBA\\Organizacion de datos\\TP\\identity_features_train.csv")
transactions = pd.read_csv("C:\\Users\\palej\\Desktop\\FIUBA\\Organizacion de datos\\TP\\transactions_train.csv")
```

```python
def ids(lista_1, lista_2):
    lista = []
    for id in lista_2:
        try:
            lista_1.remove(id)
            lista.append(True)
        except ValueError:
            lista.append(False)
    return lista
```

```python
#   Cuarto analisis
analisis = identity[ ids(transactions[ transactions["target"] == 1 ]["Id"].to_list(), identity["Id"].to_list()) ][ ["feature31", "Device", "DeviceMetadata"] ]
```

```python
analisis = analisis.fillna("Otro")
```

```python
analisis = analisis.rename(columns={"feature31":"Modelo", "Device":"Dispositivo", "DeviceMetadata":"Sistema"})
```

```python
analisis
```

```python
analisis_dispositivo = analisis["Dispositivo"].value_counts().reset_index()
```

```python
analisis_dispositivo = analisis_dispositivo.rename(columns={"count":"Cantidad"})
```

```python
analisis_dispositivo["Promedio en %"] = analisis_dispositivo["Cantidad"] * 100 / analisis_dispositivo["Cantidad"].sum()
```
