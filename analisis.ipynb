```python
import pandas as pd

formato = pd.read_csv("C:\\Users\\palej\\Desktop\\FIUBA\\Organizacion de datos\\TP\\formato_envio_alumno.csv")
identity = pd.read_csv("C:\\Users\\palej\\Desktop\\FIUBA\\Organizacion de datos\\TP\\identity_features_train.csv")
transactions = pd.read_csv("C:\\Users\\palej\\Desktop\\FIUBA\\Organizacion de datos\\TP\\transactions_train.csv")

def tipo_transferencia(numero):
    if numero == 0:
        return "Legal"
    else:
        return "Fraude"
```

```python
#   Segundo, Tercer Y Cuarto analisis
analisis = transactions[["target", "ammount", "product_category"]]
```

```python
analisis = analisis.rename(columns={"target":"Transferencia", "ammount":"Monto", "product_category":"Tipo de producto"})
```

```python
analisis["Transferencia"] = analisis["Transferencia"].map(tipo_transferencia)
```

```python
analisis = analisis.pivot_table(values="Monto", index="Tipo de producto", columns="Transferencia", aggfunc=sum)
```

```python
analisis["Total"] = analisis["Legal"] + analisis["Fraude"]
```

```python
analisis["Fraude en %"] = analisis["Fraude"] * 100 / analisis["Total"]
```

```python
analisis.describe()
```

```python
#   Sexto analisis
analisis = transactions[["target", "timestamp"]]
```

```python
analisis = analisis.rename(columns={"target":"Fraude", "timestamp":"Tiempo"})
```

```python
analisis["Transaccion"] = analisis.Fraude.map(tipo_transferencia)
```

```python
analisis = analisis.pivot_table(values="Fraude", index="Tiempo", columns="Transaccion", aggfunc=sum)
```

```python
analisis = analisis.fillna(0)
```

```python
analisis["Promedio"] = analisis["Fraude"] / analisis["Fraude"].sum()
```

```python
analisis.describe()
```
